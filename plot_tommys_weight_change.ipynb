```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_tommys_weight_change(x , y, title):
    plt.title(title)
    plt.xlabel('time')
    plt.ylabel('weight change')
    plt.yticks([])
    fig = plt.gcf()
    fig.set_size_inches(10, 8)
    plt.plot(x, y)
    plt.savefig(str(title) + '.png')
```

```python
x = np.linspace(0,3)
y = - x ** 3

plot_tommys_weight_change(x, y , title="Tommy's weight change 1")
```

```python
x = np.linspace(0,3)
y=[]

for i in x:
    if i < 1:
        y.append(- 1.5 * i)
    else:
        y.append(2 * i - 3.5)
        
y = np.array(y)

plot_tommys_weight_change(x, y , title="Tommy's weight change 2")
```

```python
x = np.linspace(0,3)
y = np.cos(15.6 * x)
 
plot_tommys_weight_change(x, y , title="Tommy's weight change 3")
```
